# 03 - Business Insights

## Executive Summary

Customer churn represents a critical revenue risk for the company.

The overall churn rate is 26.54%, indicating that more than one out of four customers discontinue service. However, the financial impact is even more significant: churned customers account for over 30% of total monthly revenue.

The analysis reveals that churn is highly concentrated in a specific segment: customers under month-to-month contracts with tenure of one year or less. This group represents approximately 25% of total revenue but contributes nearly 55% of total churn.

These findings suggest that churn is primarily an early-lifecycle issue driven by short-term contract structures. Targeted retention strategies focused on this segment could produce substantial revenue stabilization.


```python
import matplotlib.pyplot as plt
import pandas as pd

customer_churn_rate = 26.54
revenue_churn_rate = 30.50

plt.figure()
plt.bar(["Customer Churn Rate", "Revenue Churn Rate"], 
        [customer_churn_rate, revenue_churn_rate])

plt.ylabel("Percentage (%)")
plt.title("Customer Churn vs Revenue Churn")
plt.ylim(0, 40)

plt.show()
```

## Key Findings

### 1. Churn is financially disproportionate
Revenue churn exceeds customer churn, indicating that higher-paying customers are more likely to leave.

### 2. Contract type is the strongest churn driver
Month-to-month contracts show dramatically higher churn compared to one-year and two-year contracts.

```python
contract_data = pd.DataFrame({
    "Contract": ["Month-to-Month", "One Year", "Two Year"],
    "Churn Rate": [42.71, 11.27, 2.83]
})

plt.figure()
plt.bar(contract_data["Contract"], contract_data["Churn Rate"])

plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Contract Type")
plt.ylim(0, 50)

plt.show()
```

### 3. Churn is concentrated in early tenure
Customers within their first year show significantly higher churn rates than long-term customers.

```python
tenure_data = pd.DataFrame({
    "Tenure Segment": ["0-1 Years", "1-2 Years", "2-4 Years", "4+ Years"],
    "Churn Rate": [47.44, 28.71, 20.39, 9.51]
})

plt.figure()
plt.bar(tenure_data["Tenure Segment"], tenure_data["Churn Rate"])

plt.ylabel("Churn Rate (%)")
plt.title("Churn Rate by Tenure Segment")
plt.ylim(0, 50)

plt.xticks(rotation=20)

plt.show()
```

### 4. Risk is concentrated in a single high-impact segment
The combination of month-to-month contract and tenure under 12 months explains more than half of total churn.

```python
segment_revenue_share = 25.45
segment_churn_share = 54.79

plt.figure()
plt.bar(["Revenue Share", "Churn Share"], 
        [segment_revenue_share, segment_churn_share])

plt.ylabel("Percentage (%)")
plt.title("High-Risk Segment Contribution")
plt.ylim(0, 70)

plt.show()
```

## Strategic Interpretation

The data suggests that churn is not random. It follows a structural pattern tied to customer lifecycle and contract commitment.

Customers without long-term contractual commitment and limited tenure appear to be less embedded in the service ecosystem, making them more sensitive to switching costs, price perception, or early dissatisfaction.

This indicates that the primary business risk lies in onboarding quality and contract design rather than in long-term customer satisfaction.


## Recommended Actions

1. Introduce incentives for migrating month-to-month customers to annual contracts during their first year.
2. Design targeted retention campaigns between months 3 and 9, where churn risk is highest.
3. Analyze onboarding experience to identify friction points driving early churn.
4. Consider temporary discounts or bundled offers for high-value customers under short-term contracts.
